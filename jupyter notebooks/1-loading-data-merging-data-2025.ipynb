# 1. Loading Data & Light EDA

## Load and merge tables

```python
import pandas as pd
```

```python
demo_table = pd.concat([
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q1/ASCII/DEMO25Q1.txt", sep="$", low_memory=False),
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q2/ASCII/DEMO25Q2.txt", sep="$", low_memory=False),
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q3/ASCII/DEMO25Q3.txt", sep="$", low_memory=False),
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q4/ASCII/DEMO25Q4.txt", sep="$", low_memory=False),
], ignore_index=True)
```

```python
drug_table = pd.concat([
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q1/ASCII/DRUG25Q1.txt", sep="$", low_memory=False),
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q2/ASCII/DRUG25Q2.txt", sep="$", low_memory=False),
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q3/ASCII/DRUG25Q3.txt", sep="$", low_memory=False),
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q4/ASCII/DRUG25Q4.txt", sep="$", low_memory=False),
], ignore_index=True)
```

```python
reaction_table = pd.concat([
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q1/ASCII/REAC25Q1.txt", sep="$", low_memory=False),
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q2/ASCII/REAC25Q2.txt", sep="$", low_memory=False),
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q3/ASCII/REAC25Q3.txt", sep="$", low_memory=False),
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q4/ASCII/REAC25Q4.txt", sep="$", low_memory=False),
], ignore_index=True)
```

```python
outcome_table = pd.concat([
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q1/ASCII/OUTC25Q1.txt", sep="$", low_memory=False),
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q2/ASCII/OUTC25Q2.txt", sep="$", low_memory=False),
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q3/ASCII/OUTC25Q3.txt", sep="$", low_memory=False),
    pd.read_csv("raw_data/raw_data/faers_ascii_2025q4/ASCII/OUTC25Q4.txt", sep="$", low_memory=False),
], ignore_index=True)
```

### 1. light eda: show col and row num + col names

```python
# DEMO table
print("Shape of Demographics Table:", demo_table.shape)
print()
print("Summary of Demographic Table: ")
demo_table.info()
```

```python
# DRUG table
print("Shape of Drug Table:", drug_table.shape)
print()
print("Summary of Drug Table: ")
drug_table.info()
```

```python
# REAC table
print("Shape of Reaction Table:", reaction_table.shape)
print()
print("Summary of Reaction Table: ")
reaction_table.info()
```

```python
# OUTC table
print("Shape of Outcome Table:", outcome_table.shape)
print()
print("Summary of Outcome Table: ")
outcome_table.info()
```

### 3. NA values

```python
print("NA values for Demographic table:")
demo_table.isna().sum()
```

```python
print("Missing values for Drug table:")
drug_table.isna().sum()
```

```python
print("Missing values for Reaction table:")
reaction_table.isna().sum()
```

```python
print("Missing values for Outcome table:")
outcome_table.isna().sum()
```

### 3. Duplicates

```python
print("Demographic Table: ", demo_table.duplicated().sum())
print("Drug Table: ", drug_table.duplicated().sum())
print("Reaction Table: ", reaction_table.duplicated().sum())
print("Outcome Table: ", outcome_table.duplicated().sum())
```
